Question 1: Plotting a line chart of date versus temperature by adding Label on X and Y axis, and
adding a Title and Grids
to the grids: Date=[“25/12”,“26/12”,“27/12”, “28/12”,“29/12”] temp= [8.5, 10.5,6.8,7.6.6.9,5.2]

```python
import matplotlib.pyplot as plt
from datetime import datetime
import matplotlib.dates as mdates

# Sample data (replace this with your actual data)
dates = ['25/12', '26/12', '27/12', '28/12','29/12']
temperatures = [8.5,10.5,6.8,7.6,6.9]

# Convert date strings to datetime objects
dates = [datetime.strptime(date, '%d/%m') for date in dates]

# Plotting the line chart
plt.plot(dates, temperatures, marker='o', linestyle='-', color='b', label='Temperature')

# Formatting x-axis to display dates nicely
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%d/%m'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator())

# Adding labels and title
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.title('Date vs Temperature Line Chart')

# Displaying the grid
plt.grid(True)

# Adding legend
plt.legend()

# Rotating x-axis labels for better readability
plt.gcf().autofmt_xdate()

# Show the plot
plt.show()
```

Question 2: Consider the average heights and weights of persons aged 8 to 6 stored in the
following two lists: height = [121.9,124.5,129.5,134.6,139.7,147.3, 152.4, 157.5. 162.6] weight=
[19.7,21.3,23.5.25.9,28.5.32. 1,35.7,39.6,43.2]
Let us plot a line chart where:
I. X axis will represent weight.
II. Y axis will represent height.
X axis label should be “Weight in kg”,
IV. Y axis label should be “Height in cm”.
V. Colour of the line should be green.
VI. The title of the chart should be *“Average weight with respect to average height”.
VII. Line style should be dashed.
VIII. Linewidth should be 2.

```python
import matplotlib.pyplot as plt

# Sample data (replace this with your actual data)
weights = [19.7,21.3,23.5,25.9,28.5,32.1,35.7,39.6,43.2]  # Replace with actual weights
heights = [121.9,124.5,129.5,134.6,139.7,147.3,152.4,157.5,162.6]  # Replace with actual heights

# Plotting the line chart
plt.plot(weights, heights, linestyle='--', linewidth=2, color='green', label='Average Height')

# Adding labels and title
plt.xlabel('Weight')
plt.ylabel('Height')
plt.title('Average Height with respect to Weight')

# Adding legend
plt.legend()

# Display the plot
plt.show()
```

Question 3: Consider the following data:
course = [C, C++, Java, Python]
3
weight= [20,15,30,35]
Plot a bar chart and scatter plot where: IX. X axis will represent weight.
X. Y axis will represent height.
XI. X axis label should be “Courses offered”.
XII. Y axis label should be ” No. of students enrolled”.
XIII. Colour of the bar plot should be maroon.
XIV. XV. Colour of the markers for the scatter plot should be blue. The title of the chart should
be “Students enrolled in different courses”

```python
import matplotlib.pyplot as plt
# Data
courses = ['C', 'C++', 'Java', 'Python']
weights = [20, 15, 30, 35]
# Bar chart
plt.figure(figsize=(10, 5))
plt.bar(courses, weights, color='maroon')
plt.xlabel('Courses offered')
plt.ylabel('No. of students enrolled')
plt.title('Students enrolled in different courses (Bar Chart)')
plt.show()
# Scatter plot
plt.figure(figsize=(10, 5))
plt.scatter(weights, courses, color='blue', marker='o')
plt.xlabel('Courses offered')
plt.ylabel('No. of students enrolled')
plt.title('Students enrolled in different courses (Scatter Plot)')
plt.show()
```
